```python
import os
import geopy
from geopy import distance
import csv

output_dir = "output"
output_file = "cities_distance.csv"
output_path = os.path.join(output_dir, output_file)

home_country = "India"
home_city = "Shimla"

# Find the coordinates of the home city
home_city_coord = None
with open(path, 'r', encoding='utf-8') as file:
    csv_reader = csv.DictReader(file, delimiter=',')
    for row in csv_reader:
        if row['city_ascii'] == home_city:
            home_city_coord = (row['lat'], row['lng'])
            break

if home_city_coord:
    with open(output_path, 'w', newline='', encoding='utf-8') as outfile:
        fieldnames = ['city', 'distance']
        csv_writer = csv.DictWriter(outfile, fieldnames=fieldnames)
        csv_writer.writeheader()

        with open(path, 'r', encoding='utf-8') as infile:
            csv_reader = csv.DictReader(infile, delimiter=',')
            for row in csv_reader:
                if row['country'] == home_country and row["city_ascii"] != home_city:
                    city_coord = (row['lat'], row['lng'])
                    try:
                        distance_km = distance.geodesic(home_city_coord, city_coord).km
                        csv_writer.writerow(
                            {
                                'city': row['city_ascii'],
                                'distance': distance_km
                            }
                        )
                    except ValueError:
                        # Handle cases where coordinates are invalid
                        print(f"Skipping {row['city_ascii']} due to invalid coordinates: {city_coord}")
else:
    print(f"Home city '{home_city}' not found in the dataset.")
```
